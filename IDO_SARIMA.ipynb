```python
import numpy as np
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX

train_size = int(len(ind.loc[:'2022-06-01']))
train_data, test_data = ind['ConfirmedCases'][:train_size], ind['ConfirmedCases'][train_size:]

# Set the range of orders for p, d, q, P, D, Q, and s
p_range = range(0, 3)  # For example, from 0 to 2
d_range = range(0, 2)  # Use 1 for first-order differencing, or 0 for no differencing
q_range = range(0, 3)
P_range = range(0, 2)
D_range = range(0, 2)
Q_range = range(0, 3)
s = 12  # Seasonality period

# Create a list to store the best models along with orders
best_models_aic = []
best_models_bic = []

# Iterate over all possible combinations of orders
for p in p_range:
    for d in d_range:
        for q in q_range:
            for P in P_range:
                for D in D_range:
                    for Q in Q_range:
                        try:
                            # Fit the SARIMA model
                            sarima_model = SARIMAX(train_data, order=(p, d, q), seasonal_order=(P, D, Q, s))
                            sarima_fit = sarima_model.fit()

                            # Store the model along with AIC and BIC
                            best_models_aic.append(((p, d, q, P, D, Q, s), sarima_fit, sarima_fit.aic))
                            best_models_bic.append(((p, d, q, P, D, Q, s), sarima_fit, sarima_fit.bic))

                        except:
                            continue

# Sort the models based on AIC and BIC
best_models_aic.sort(key=lambda x: x[2])
best_models_bic.sort(key=lambda x: x[2])

# Select the top 5 models based on AIC and BIC
top_5_aic = best_models_aic[:5]
top_5_bic = best_models_bic[:5]

# Print the selected models and their AIC/BIC along with orders
for i, (orders, model, aic) in enumerate(top_5_aic):
    print(f"Top AIC Model {i+1}: AIC = {aic}")
    print(f"Orders: p={orders[0]}, d={orders[1]}, q={orders[2]}, P={orders[3]}, D={orders[4]}, Q={orders[5]}, s={orders[6]}")
    print(model.summary())
    print("=" * 50)

for i, (orders, model, bic) in enumerate(top_5_bic):
    print(f"Top BIC Model {i+1}: BIC = {bic}")
    print(f"Orders: p={orders[0]}, d={orders[1]}, q={orders[2]}, P={orders[3]}, D={orders[4]}, Q={orders[5]}, s={orders[6]}")
    print(model.summary())
    print("=" * 50)
```

```python
ind['Diff']=ind['ConfirmedCases']-ind['ConfirmedCases'].shift(1)

train_size = int(len(ind.loc[:'2022-06-01']))
train_data, test_data = ind['ConfirmedCases'][:train_size], ind['ConfirmedCases'][train_size:]

sarima_b = SARIMAX(train_data, order=(1, 1, 2), seasonal_order=(1, 1, 1, 12))
sarima_fit = sarima_b.fit()

f_steps = len(test_data)
forecasted_values = sarima_fit.forecast(steps=f_steps)
```

```python
from sklearn.metrics import mean_absolute_error
import numpy as np

def calculate_mape(y_true, y_pred):
    """
    Calculate Mean Absolute Percentage Error (MAPE).

    Parameters:
        y_true (numpy array): True values.
        y_pred (numpy array): Predicted values.

    Returns:
        float: MAPE in percentage notation.
    """
    # Calculate the MAE (Mean Absolute Error)
    mae = mean_absolute_error(y_true, y_pred)

    # Calculate the MAPE
    mape = (mae / np.mean(y_true)) * 100

    return mape

mape_result = calculate_mape(test_data, forecasted_values)
print(f"MAPE: {mape_result:.2f}%")
```
